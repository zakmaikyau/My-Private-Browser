```python
from ase.build import graphene, molecule

# 1. Create a 4x4 Graphene Supercell
# Size (4,4,1) gives us enough room so the CO2 doesn't "see" itself in the next cell
system = graphene(formula='C2', a=2.46, size=(4, 4, 1), vacuum=10)

# 2. Create the Pyridinic Nitrogen Defect
# Remove one Carbon (the vacancy)
del system[0]
# Replace its neighbor with Nitrogen
system[0].symbol = 'N'

# 3. Add the CO2 "Guest"
guest = molecule('CO2')
guest.rotate(90, 'x') # Orientation matters for bonding
# Position it 2.5 Angstroms above the Nitrogen dopant
guest.translate([1.2, 0.7, 2.5])

system = system + guest
print(f"✅ Step 2: N-Graphene + CO2 built ({len(system)} atoms).")
```

```python
from gpaw import GPAW, PW
from ase.optimize import BFGS

# Setup the high-reliability calculator
calc = GPAW(mode='lcao',
            basis='dzp',
            kpts=(2, 2, 1),
            txt='total_system.log')
system.calc = calc

# Run the relaxation
relax = BFGS(system, trajectory='n_graphene_co2.traj')
print("Starting relaxation... (This may take 15-20 minutes)")
relax.run(fmax=0.08)

e_total = system.get_potential_energy()
print(f"✅ Step 3 Complete! Total Energy: {e_total:.4f} eV")
```

```python
# Part A: Lone CO2 (We use the standard value we found earlier)
e_co2 = -22.7036

# Part B: The N-Graphene Slab (without CO2)
slab_only = system[:-3].copy()
slab_only.calc = GPAW(mode='lcao', basis='dzp', kpts=(2, 2, 1), txt='slab_only.log')
e_slab = slab_only.get_potential_energy()

# The Final Result
e_ads = e_total - (e_slab + e_co2)
print(f"--- RESULTS ---")
print(f"Adsorption Energy: {e_ads:.4f} eV")
```

```python
# Identify atoms
c_idx = [a.index for a in system if a.symbol == 'C'][-1] # The CO2 carbon
o_indices = [a.index for a in system if a.symbol == 'O']
n_idx = [a.index for a in system if a.symbol == 'N'][0]

# Calculate metrics
angle = system.get_angle(o_indices[0], c_idx, o_indices[1])
dist = system.get_distance(c_idx, n_idx)

print(f"Final O-C-O Angle: {angle:.2f}°")
print(f"C-N Distance: {dist:.3f} Å")

if angle < 178:
    print("🚀 SUCCESS: The CO2 is bent! This shows chemical activation.")
```

```python
# 1. Manually move the CO2 Carbon to exactly 1.8A above the Nitrogen
# This is the 'Chemisorption' zone
c_idx = [a.index for a in system if a.symbol == 'C'][-1]
n_idx = [a.index for a in system if a.symbol == 'N'][0]

# Move the whole molecule down
target_height = system[n_idx].position[2] + 1.8
current_height = system[c_idx].position[2]
shift = target_height - current_height

# Move the Carbon and its two Oxygens
for i in [c_idx, o_indices[0], o_indices[1]]:
    system[i].position[2] += shift

print(f"Manual Reset: Carbon is now {system.get_distance(c_idx, n_idx):.2f} Å from Nitrogen.")

# 2. Re-run relaxation with a tighter focus
calc = GPAW(mode='lcao', basis='dzp', kpts=(2, 2, 1), txt='forced_n_graphene.log')
system.calc = calc
relax = BFGS(system, trajectory='forced_n.traj')
relax.run(fmax=0.08)

# 3. Final Check
new_e = system.get_potential_energy()
new_e_ads = new_e - (e_slab + e_co2)
new_angle = system.get_angle(o_indices[0], c_idx, o_indices[1])

print(f"--- THE 'FORCED' RESULTS ---")
print(f"New Adsorption Energy: {new_e_ads:.4f} eV")
print(f"New O-C-O Angle: {new_angle:.2f}°")
```

```python
# Check the C-O bonds in the captured state
bond1 = system.get_distance(c_idx, o_indices[0])
bond2 = system.get_distance(c_idx, o_indices[1])

print(f"--- ACTIVATION EVIDENCE ---")
print(f"C-O Bond 1: {bond1:.3f} Å")
print(f"C-O Bond 2: {bond2:.3f} Å")
print(f"Average Stretch: {((bond1+bond2)/2 - 1.16):.3f} Å")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Calculate the DOS from the calculator
# 'npts' is the number of points, 'width' is the broadening (0.1 is standard)
energies, dos = calc.get_dos(spin=0, npts=2001, width=0.1)
fermi_level = calc.get_fermi_level()

# 2. Adjust energies so that the Fermi Level is at 0 (Research Standard)
energies -= fermi_level

# 3. Create the plot
plt.figure(figsize=(8, 5))
plt.plot(energies, dos, color='blue', linewidth=1.5, label='Total DOS')
plt.fill_between(energies, dos, color='blue', alpha=0.2)

# Mark the Fermi Level
plt.axvline(x=0, color='red', linestyle='--', label='Fermi Level ($E_F$)')

plt.xlim(-10, 5) # Focus on the bonding region
plt.ylim(0, max(dos)*1.1)
plt.xlabel('Energy - $E_F$ (eV)')
plt.ylabel('Density of States (states/eV)')
plt.title('Electronic Structure of $CO_2$ Adsorbed on N-Graphene')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Identify the indices
co2_indices = [len(system)-3, len(system)-2, len(system)-1]
n_idx = [atom.index for atom in system if atom.symbol == 'N'][0]

# 2. Calculate projected DOS (Notice 'atoms=' is removed)
energies, dos_co2 = calc.get_all_electron_ldos(co2_indices, spin=0, npts=2001, width=0.1)
energies, dos_n = calc.get_all_electron_ldos([n_idx], spin=0, npts=2001, width=0.1)

# Normalize energies to Fermi Level (Standard for publications)
energies -= calc.get_fermi_level()

# 3. Plotting
plt.figure(figsize=(10, 6))
plt.plot(energies, dos_co2, label='$CO_2$ Projection', color='forestgreen', lw=1.5)
plt.plot(energies, dos_n, label='Nitrogen Projection', color='darkorange', lw=1.5)

plt.fill_between(energies, dos_co2, color='forestgreen', alpha=0.3)
plt.fill_between(energies, dos_n, color='darkorange', alpha=0.3)

plt.axvline(x=0, color='red', linestyle='--', label='Fermi Level ($E_F$)')
plt.title('PDOS Analysis: Orbital Hybridization at the Active Site', fontsize=12)
plt.xlabel('Energy - $E_F$ (eV)')
plt.ylabel('Density of States')
plt.xlim(-8, 2)
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
